## 50 most popular papers in IEEE Transactions on Affective Computing up to February 2019

```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```python
df_slr = pd.read_excel(
    './Papers_IEEE_last_years.xlsx', 
    'popular_papers_february_2019', 
    dtype={
        'Year_Date_Publication': np.intc,
        'Cited by': np.intc,
    }
)
df_slr
```

```python
pages = df_slr['Pages']
plt.hist(pages)
plt.title('Number of pages (in ascendent order)')
plt.xlabel('Number of pages')
print('Mean(number of pages) = {}'.format(pages.mean()))
print('SD(number of pages) = {}'.format(pages.std()))
```

```python
df_participants = df_slr['Participants']
df_participants.replace(np.nan, 0, inplace=True)
```

```python
plt.barh(np.arange(len(df_participants)), df_participants.sort_values(inplace=False))
plt.title('Number of participants in evaluations (in ascendent order)')
plt.xlabel('Number of participants')
print('Mean(number of participants) = {}'.format(df_participants.mean()))
print('SD(number of participants) = {}'.format(df_participants.std()))
```

```python
plt.hist(df_participants, bins=10, align='mid', color='orange')
plt.title('Distribución de participantes en evaluaciones')
matplotlib.rcParams.update({'font.size': 12})

plt.xticks(np.arange(min(df_participants), max(df_participants)+1, 6.0))
plt.xlabel('Number of people in each evaluation group')
plt.show()
```

```python
unique_elements, counts_elements = np.unique(df_slr['Year_Date_Publication'], return_counts=True)
year_count = pd.DataFrame({'Año': unique_elements, 'Apariciones': counts_elements}, columns=['Año', 'Apariciones'])
year_count.set_index('Año', inplace=True)
plt.figure(figsize=(12, 6))
plt.plot(np.arange(len(unique_elements)), counts_elements, marker='o', color='orange')
plt.xticks(np.arange(len(unique_elements)), unique_elements)
plt.ylim(0, 7)
plt.grid(True)
plt.title('Papers per year')
plt.show()
```

```python
df_slr['Keywords'] = df_slr['Keywords'].map(lambda keywords: keywords.lower().split(','))
df_slr['Keywords'] = df_slr['Keywords'].map(lambda a: list(map(lambda b: b.strip(), a)))
df_slr['Keywords_cited_by'] = df_slr.apply(lambda row: list(map(lambda kw: (row['Cited by'], kw), row['Keywords'])), axis=1)
```

```python
kw_list = np.char.strip(np.concatenate(df_slr['Keywords']))
unique_kw = np.unique(kw_list)
unique_kw
```

```python
from functools import reduce
keyword_cited_ll = df_slr['Keywords_cited_by'].values.tolist()
keyword_cited_list = reduce(lambda a, b: a + b, keyword_cited_ll, [])
keyword_frecuencies = {kw:0 for kw in unique_kw} 
```

```python
for n_cites, key in keyword_cited_list:
    keyword_frecuencies[key] += n_cites
keyword_frecuencies
```

```python
from wordcloud import WordCloud

wordcloud = WordCloud(width=800, height=400, background_color="white").generate_from_frequencies(keyword_frecuencies)

# Display the generated image:
fig = plt.figure(figsize=(10,6))
plt.imshow(wordcloud, interpolation='kaiser')
plt.axis("off")
plt.show()
```

```python
from PIL import Image
brain_mask = np.array(Image.open("brain.png"))
transformed_brain_mask = np.ndarray((brain_mask.shape[0],brain_mask.shape[1]), np.int32)

def transform_format(val):
    if val == 0:
        return 255
    else:
        return val

for i in range(len(brain_mask)):
    transformed_brain_mask[i] = list(map(transform_format, brain_mask[i]))
```

```python
wordcloud = WordCloud(
    width=1000, 
    height=600, 
    mask=transformed_brain_mask, 
    contour_width=3, 
    contour_color='palevioletred', 
    background_color='white'
)

wordcloud.generate_from_frequencies(keyword_frecuencies)

# Display the generated image:
fig = plt.figure(figsize=(20,10))
plt.imshow(wordcloud, interpolation='kaiser')
plt.axis("off")
plt.show()
```
