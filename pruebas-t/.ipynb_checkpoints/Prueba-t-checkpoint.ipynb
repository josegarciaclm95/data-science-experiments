```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df_eval = pd.read_csv("datos_evaluacion_CSV.csv", sep=';')
df_eval.loc[np.isnan(df_eval['help']), 'help'] = 0.0
df_eval
```

```python
from scipy import stats

def t_test(dataA, dataB, comparedData, hypothesis):
    N = len(dataA)
    ## Calculate the Standard Deviation
    #Calculate the variance to get the standard deviation
    #For unbiased max likelihood estimate we have to divide the var by N-1, and therefore the parameter ddof = 1
    var_a = dataA.var(ddof=1)
    var_b = dataB.var(ddof=1)

    #std deviation
    s = np.sqrt((var_a + var_b)/2)
    ## Calculate the t-statistics
    t = (dataA.mean() - dataB.mean())/(s*np.sqrt(2/N))
    ## Compare with the critical t-value
    #Degrees of freedom
    df = 2*N - 2
    print('*****************T-TEST****************************')
    print('Grados de libertad: {}'.format(df))
    #p-value after comparison with the t 
    p = 1 - stats.t.cdf(t,df=df)
    '''
    print("t = " + str(t))
    print("p = " + str(2*p))
    print("Null hypothesis: {}".format(hypothesis))
    
    if(t < p):
        print("Null hypothesis is TRUE for the data '{}'".format(comparedData))
    else:
        print("Null hypothesis is FALSE for the data '{}'".format(comparedData))
    '''
    ## Cross Checking with the internal scipy function
    t2, p2 = stats.ttest_ind(dataA,dataB)
    print("t = " + str(t2))
    print("p = " + str(2*p2))
    if(t2 < p2):
        print("(CROSS CHECKING) Null hypothesis is TRUE for the data '{}'".format(comparedData))
    else:
        print("(CROSS CHECKING) Null hypothesis is FALSE for the data '{}'".format(comparedData))
    print('*****************T-TEST****************************')
```

```python
for col in ['attempts', 'time', 'mistakes', 'strokes', 'excessive']:
    t_test(
        df_eval.loc[df_eval.affective == 'SI'][col], 
        df_eval.loc[df_eval.affective == 'NO'][col], 
        col, 
        'Average of {} are not statistically different'.format(col)
    )
    plt.figure(figsize=(6, 4))
    for cnt in range(5):
        plt.plot(np.arange(len(df_eval.loc[df_eval.affective == 'SI'])), df_eval.loc[df_eval.affective == 'SI'][col], c='r')
        plt.plot(np.arange(len(df_eval.loc[df_eval.affective == 'NO'])), df_eval.loc[df_eval.affective == 'NO'][col], c='b')
    plt.tight_layout()
    plt.show()
```
